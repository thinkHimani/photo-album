## Add data to opensearch

```python
import requests
import uuid
import json
import boto3
```

```python
host = 'https://search-cc-photos-qco2o44jvnbkzdh5q6cycer6l4.us-east-1.es.amazonaws.com'
_index = 'photos'
_type = 'Photo'
_id = uuid.uuid1().hex

path = '/{}/{}/{}/'.format(_index, _type, _id)
master_user = "test"
master_password = "Test@1234"

payload = {'objectKey': 'landscape_resized.jpg', 'bucket': 'cc-b2', 'createdTimestamp': '2022-04-02 23:26:14.305358', 'labels': ['City','Aniket','lol']}

url = host + path
r = requests.post(url, auth=(master_user, master_password), json=payload)
print(r.content)
```

## Create slot type

```python
import pandas as pd
import json
```

```python
labels_df = pd.read_csv('data/labels.csv', header = None)
labels_df[0].values
```

```python
cust_slot = {
    "enumerationValues": [],
    "name": "LabelTypes",
    "description": "Labels of photos"
}
```

```python
for label in labels_df[0].values:
    temp = {}
    temp["value"] = label
    cust_slot['enumerationValues'].append(temp)
```

```python
cust_slot
```

```python
with open('data/slots.json', 'w') as fp:
    json.dump(cust_slot, fp)
```
